```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("rotskoff_notebook.mplstyle")
```

```python
def fel(a,b,c,x):
   return a*x**2 + b*x**4 + c*x**3
x = np.linspace(-2.5, 2.5, 1200)
plt.plot(x, fel(-2,3,2,x), lw=4,label='Protein', color='red')
plt.plot(x, fel(-2,3,0,x), lw=4,label='Water', color='blue')
plt.xlabel(r"$x$ (ab. units)", fontsize=20) 
plt.ylabel(r"$\mathcal{F}(x)$ (ab. units)", fontsize=20)
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.legend(loc='upper center', fontsize=15)
plt.tick_params(labelsize=15)
plt.xticks([-1.0, -0.5, 0.0, 0.5, 1.0]) 
plt.yticks([-1.0, -0.5, 0.0, 0.5, 1.0])
plt.savefig("../reports/free_energy_landscape.png", dpi=300)
```

```python
x = np.linspace(-2.5, 2.5, 1200)
plt.plot(x, fel(-2,3,2,x), lw=3,label='low $pH$', color='#008080')
plt.plot(x, fel(-2,3,0,x), lw=3,label='mid $pH$', color='#7B3294')
plt.plot(x, fel(-2,3,-2,x), lw=3,label='high $pH$', color='#E69F00')
plt.xlabel(r"$x$ (ab. units)", fontsize=20) 
plt.ylabel(r"$\mathcal{F}(x)$ (ab. units)", fontsize=20)
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.legend(loc='upper center', fontsize=15)
plt.tick_params(labelsize=15)
plt.xticks([-1.0, -0.5, 0.0, 0.5, 1.0]) 
plt.yticks([-1.0, -0.5, 0.0, 0.5, 1.0])
plt.savefig("../reports/free_energy_landscape_pH.png", dpi=300)
```
